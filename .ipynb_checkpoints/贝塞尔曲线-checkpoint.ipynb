```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from scipy.special import comb

def getParam(n, points):
    params = []
    l = len(points)
    if n != l - 1:
        print('Points number is wrong!')
        return None
    for i in range(l):
        tmp = np.array([0, 0], dtype = np.float64)
        for j in range(l-i):
            tmp += comb(n-i, j) * points[j] * (-1) ** (n-i-j)
        tmp *= comb(l-1, i)
        params.append(tmp)
    return params
```

```python
%matplotlib inline

fig1 = plt.figure(figsize = (20, 20))
points = np.array([[0,0], [100, 150], [200, 150], [300, 0]], dtype= np.float64)
plt.plot([p[0] for p in points], [p[1] for p in points], 'ko--')

t1 = np.arange(0.0, 1.0, 0.01)
params = getParam(3, points)

def xt(params, t):
    l = len(params)
    ans = 0
    for i in range(l):
        ans += params[i][0] * t ** (l-i-1)
    return ans

def yt(params, t):
    l = len(params)
    ans = 0
    for i in range(l):
        ans += params[i][1] * t ** (l-i-1)
    return ans

plt.plot(xt(params, t1), yt(params, t1), 'r-')
plt.axis([0, 400, 0, 400])
plt.show()
```

```python
fig2 = plt.figure(figsize = (20, 20))

points = np.array([[0,0], [100, 150], [200, 150], [300, 0]], dtype= np.float64)

params = getParam(3, points)
plt.plot(xt(params, t1), yt(params, t1), 'r-')

vec = [300, 0]
points = points - vec
params = getParam(3, points)
plt.plot(xt(params, t1), yt(params, t1), 'g-')

plt.axis([-400, 400, 0, 400])
plt.show()
```

```python
fig = plt.figure(figsize = (15, 15))
points = np.array([[20,0], [20, 11.04], [11.04, 20], [0, 20]], dtype= np.float64)
points1 = np.array([[20,0], [20, 20], [0, 20]])

plt.plot([p[0] for p in points], [p[1] for p in points], 'ko--')
plt.plot([p[0] for p in points1], [p[1] for p in points1], 'bo--')

t1 = np.arange(0.0, 1.0, 0.01)

params = getParam(3, points)
params1 = getParam(2, points1)

def xt(params, t):
    l = len(params)
    ans = 0
    for i in range(l):
        ans += params[i][0] * t ** (l-i-1)
    return ans

def yt(params, t):
    l = len(params)
    ans = 0
    for i in range(l):
        ans += params[i][1] * t ** (l-i-1)
    return ans

plt.plot(20 * np.sin(np.pi * (1-t1) / 2), 20 * np.cos(np.pi * (1-t1) / 2), 'b-')
plt.plot(xt(params1, t1), yt(params1, t1), 'y-')
plt.plot(xt(params, t1), yt(params, t1), 'r-')
plt.show()
```
