## 例1
由$P_0(0, 0, 0)、P_1(10, -10, 0)、P_2(30, -20, 0)、P_3(40, -10, 0) 、P_4(40, 10, 0)、P_5(30, 20, 0)、P_6(10, 10, 0)$确定一条均匀三次B样条曲线，计算曲线第3子段上点$p_3(0.5)$的值。

```python
import numpy as np
from scipy.special import comb

points = np.array([[0,0,0], [10,-10,0], [30,-20,0], [40, -10, 0], [40, 10, 0], [30, 20, 0], [10, 10, 0]])
print(points.shape)
```

### 三阶B样条曲线定义为：

$$
Q_{i+1}(u) = \sum_{j=0}^{3}B_{j,3}(u)P_{i+j} = [u^3, u^2, u, 1]\frac{1}{6} \left[ \begin{matrix} -1 &3 &-3 &1\\ 3 &-6 &3 &0\\ -3 &0 &3 &0 \\ 1 &4 &1 &0 \end{matrix}\right]\left[ \begin{matrix} P_i\\ P_{i+1}\\ P_{i+2} \\ P_{i+3} \end{matrix}\right]
$$

```python
def Qn(n, u, points):
    P = points[n-1:n+3, :]
    U = np.array([[u ** 3, u ** 2, u ** 1, 1]])
    M = np.array([[-1, 3, -3, 1], [3, -6, 3, 0], [-3, 0, 3, 0], [1, 4, 1, 0]])
    return np.dot(np.dot(U, M), P) / 6
```

```python
p = Qn(3, 0.5, points)
print(p)
```

### 创建三阶B样条曲线

```python
def makespline3(points):
    l = len(points)
    uArray = np.arange(0.0, 1.0, 0.01)
    ans = np.array([[0,0,0]])
    for i in range(l-3):
        P = points[i:i+4, :]
        for u in uArray:
            U = np.array([[u ** 3, u ** 2, u ** 1, 1]])
            M = np.array([[-1, 3, -3, 1], [3, -6, 3, 0], [-3, 0, 3, 0], [1, 4, 1, 0]])
            tmp = np.dot(np.dot(U, M), P) / 6
            ans = np.concatenate((ans, tmp), axis=0)
    return ans[1:,:]
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,10))
plt.plot(points[:,0], points[:,1], 'ko--')

line = makespline3(points)
plt.plot(line[:,0], line[:,1], 'r-')

plt.axis([-50, 50, -50, 50])
plt.show()
```

## 例2

已知均匀双三次B样条曲面P(u,v)由以下控制点定义：
$$
\begin{matrix}
P_{00}(0, 0, 0) & P_{01}(20, 0, 20)  &P_{02}(40, 0, 20)  &P_{03}(60, 0, 0)    &P_{04}(80, 0, -10)\\
P_{10}(0, 20, 20) & P_{11}(20, 20, 40) &P_{12}(40, 20, 40) &P_{13}(60, 20, 20)   &P_{14}(80, 20, 0)\\
P_{20}(0, 40, 20) & P_{21}(20, 40, 40) &P_{22}(40, 40, 40) &P_{23}(60, 40, 20)   &P_{24}(80, 40, 0)\\
P_{30}(0, 60, 0) & P_{31}(20, 60, 20)  &P_{32}(40, 60, 20) &P_{33}(60, 60, 0)    &P_{34}(80, 60, -10)\\
P_{40}(0, 80, -20) & P_{41}(20, 80, 0)  &P_{42}(40, 80, 0)  &P_{43}(60, 80, -20)  &P_{44}(80, 80, 0)
\end{matrix}
$$
求该B样条曲面上u向排列第2， v向排列第1的子曲面片上点$P(0.5,0.5)$的值。

```python
points = np.array([[[0,0,0], [20,0,20], [40,0,20], [60,0,0], [80,0,-10]],\
                  [[0,20,20], [20,20,40], [40,20,40], [60,20,20], [80,20,0]],\
                  [[0,40,20], [20,40,40], [40,40,40], [60,40,20], [80,40,0]],\
                  [[0,60,0], [20,60,20], [40,60,20], [60,60,0], [80,60,-10]],\
                  [[0,80,-20], [20,80,0], [40,80,0], [60,80,-20],[80,80,0]]])
```

```python
def Bpoints(u, v, m, n, points):
    P = points[m:m+4, n:n+4, :]
    U = np.array([[u ** 3, u ** 2, u ** 1, 1]])
    M = np.array([[-1, 3, -3, 1], [3, -6, 3, 0], [-3, 0, 3, 0], [1, 4, 1, 0]])
    V = np.array([[v ** 3, v ** 2, v ** 1, 1]])
    return np.dot(np.dot(U, M)/6, np.dot(np.dot(V, M), P)/ 6)
```

```python
print(Bpoints(0.5, 0.5, 1, 0, points))
```

```python
def Bsurface(points):
    uArray = np.arange(0.0, 1.0, 0.01)
    vArray = np.arange(0.0, 1.0, 0.01)
    uArray, vArray = np.meshgrid(uArray, vArray)
    uArray = uArray.flatten()
    vArray = vArray.flatten()
    ans = None
    m, n = points.shape[:2]
    for i in range(m-3):
        for j in range(n-3):
            P = points[i:i+4, j:j+4, :]
            for u, v in zip(uArray, vArray):
                U = np.array([[u ** 3, u ** 2, u ** 1, 1]])
                M = np.array([[-1, 3, -3, 1], [3, -6, 3, 0], [-3, 0, 3, 0], [1, 4, 1, 0]])
                V = np.array([[v ** 3, v ** 2, v ** 1, 1]])
                current = np.dot(np.dot(V, M)/6, np.dot(np.dot(U, M), P)/ 6).reshape(1, 3)
                ans = current if not isinstance(ans, np.ndarray) else np.concatenate((ans, current), axis=0)
    return ans
```

```python
surface = Bsurface(points)
#print(surface.shape)
#print(surface[:5])
```

```python
def getReal(aline):
    X1 = aline[0:10000].reshape(100, 100)
    X2 = aline[10000:20000].reshape(100, 100)
    X3 = aline[20000:30000].reshape(100, 100)
    X4 = aline[30000:40000].reshape(100, 100)
    X_Common1 = np.concatenate((X1, X2), axis=1)
    X_Common2 = np.concatenate((X3, X4), axis=1)
    return np.concatenate((X_Common1, X_Common2), axis=0)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig2 = plt.figure(figsize=(10,10))
ax3= Axes3D(fig2)
X = getReal(surface[:,0])
Y = getReal(surface[:,1])
Z = getReal(surface[:,2])
ax3.plot_surface(X,Y,Z,cmap='rainbow')
ax3.scatter(points[:,:,0], points[:,:,1], points[:,:,2], c='k', label='control points')
 
#ax3.set_xlim(-20, 60)
#ax3.set_ylim(0, 80)
#ax3.set_zlim(-20, 60)
ax3.set_xlabel('X', fontdict={'size': 15, 'color': 'black'})
ax3.set_ylabel('Y', fontdict={'size': 15, 'color': 'black'})
ax3.set_zlabel('Z', fontdict={'size': 15, 'color': 'black'})
plt.show()
```

## 例3

已知均匀双三次B样条曲面P(u,v)由以下控制点定义：
$$
\begin{matrix}
P_{00}(0, 0, 0) & P_{01}(20, 0, 20)  &P_{02}(40, 0, 20)  &P_{03}(60, 0, 0)    &P_{04}(80, 0, -10)\\
P_{10}(0, 20, 20) & P_{11}(20, 20, 40) &P_{12}(40, 20, 40) &P_{13}(60, 20, 20)   &P_{14}(80, 20, 0)\\
P_{20}(0, 40, 20) & P_{21}(20, 40, 40) &P_{22}(40, 40, 40) &P_{23}(60, 40, 20)   &P_{24}(80, 40, 0)\\
P_{30}(0, 60, 0) & P_{31}(20, 60, 20)  &P_{32}(40, 60, 20) &P_{33}(60, 60, 0)    &P_{34}(80, 60, -10)
\end{matrix}
$$
试设计构造一条Bezier曲线与曲面上的曲线$P(0.5,v)$的终点$Q(0.5,1.0)$一阶光滑连接

```python
points = np.array([[[0,0,0], [20,0,20], [40,0,20], [60,0,0], [80,0,-10]],\
                  [[0,20,20], [20,20,40], [40,20,40], [60,20,20], [80,20,0]],\
                  [[0,40,20], [20,40,40], [40,40,40], [60,40,20], [80,40,0]],\
                  [[0,60,0], [20,60,20], [40,60,20], [60,60,0], [80,60,-10]]])

surface = Bsurface(points)
```

```python
def getReal(aline):
    X1 = aline[0:10000].reshape(100, 100)
    X2 = aline[10000:20000].reshape(100, 100)
    return np.concatenate((X1, X2), axis=1)
```

```python
def getBline(u,points):
    vArray = np.arange(0, 1.01, 0.01)
    ans = None
    m, n = points.shape[:2]
    for i in range(m - 3):
        for j in range(n - 3):
            for v in vArray:
                c = Bpoints(u, v, i, j, points)
                ans = c if not isinstance(ans, np.ndarray) else np.concatenate((ans, c), axis=0)
    return ans
```

```python
fig2 = plt.figure(figsize=(10,10))
ax3= Axes3D(fig2)

## 绘制三次B样条去曲面，
X = getReal(surface[:,0])
Y = getReal(surface[:,1])
Z = getReal(surface[:,2])
ax3.plot_surface(X,Y,Z,cmap='rainbow')

## 绘制曲线P(0.5,v)
pv = getBline(0.5, points)
ax3.plot(pv[:,:,0].flatten(), pv[:,:,1].flatten(),pv[:,:,2].flatten(), color='black')

## 绘制终点Q(0.5,1.0)
q = Bpoints(0.5, 1.0, 0, 1, points)
print(q)
ax3.scatter(q[:,:,0], q[:,:,1], q[:,:,2])

## 通过曲线P反求控制点，计算出Q处的一阶切矢

## 根据Q即为Bezier曲线的起点P0，根据一阶切矢确定第二个控制点P1，P2,P3可任意设定，不影响Beizer曲线与均匀三次B样条曲面的一阶相切。

ax3.set_xlabel('X', fontdict={'size': 15, 'color': 'black'})
ax3.set_ylabel('Y', fontdict={'size': 15, 'color': 'black'})
ax3.set_zlabel('Z', fontdict={'size': 15, 'color': 'black'})
plt.show()
```

## 例4

如图，已知一均匀双三次B样条曲面$Q(u,v)$由以下控制点（$Q_{00}, \cdots, Q_{34}$）定义，另一双三次Bezier样条曲面$P(t,w)$由控制点$(P_{00},\cdots, P_{33})$
定义，试分别设计一均匀三次B样条曲面及三次Beizier曲面使前述二曲
面光滑连接。


```python
## 三次Bezier曲面控制点pPoints
pPoints = np.array([[[0,50, 50],[100,50, 100],[100,50, 150],[0,50, 200]],\
                   [[100,100,50],[200,100,100],[200,100,150],[100,100,200]],\
                   [[100,150,50],[200,150,100],[200,150,150],[100,150,200]],\
                   [[0,200,50],[100,200,100],[100,200,150],[0,200,200]]])
points = np.array([[[0,0,0], [20,0,20], [40,0,20], [60,0,0], [80,0,-10]],\
                  [[0,20,20], [20,20,40], [40,20,40], [60,20,20], [80,20,0]],\
                  [[0,40,20], [20,40,40], [40,40,40], [60,40,20], [80,40,0]],\
                  [[0,60,0], [20,60,20], [40,60,20], [60,60,0], [80,60,-10]]])
```

```python
## 绘制Beizer曲面函数
def puv(u,v, points):
    bu = np.array([comb(3,0) * (1 - u) ** 3, comb(3,1) * u * (1 - u) ** 2, comb(3, 2) * u ** 2 * (1-u), comb(3, 3) * u ** 3])
    bv = np.array([comb(3,0) * (1 - v) ** 3, comb(3,1) * v * (1 - v) ** 2, comb(3, 2) * v ** 2 * (1-v), comb(3, 3) * v ** 3])
    return np.dot(np.dot(bu, points).T, bv)

def getBsurface(points):
    u = np.arange(0.0, 1.0, 0.01)
    v = np.arange(0.0, 1.0, 0.01)
    u, v = np.meshgrid(u, v)
    u = u.flatten()
    v = v.flatten()
    ans = None
    for i in range(10000):
        p = puv(u[i], v[i], points).reshape(1,3)
        ans = np.concatenate((ans, p), axis=0) if isinstance(ans, np.ndarray) else p
    return ans
```

```python
fig2 = plt.figure(figsize=(10,10))
ax3= Axes3D(fig2)

##绘制3次Bezier曲面
ans = getBsurface(pPoints)
X = ans[:,0].reshape(100, 100)
Y = ans[:,1].reshape(100, 100)
Z = ans[:,2].reshape(100, 100)
ax3.plot_surface(X,Y,Z,cmap='rainbow')
 
## 绘制双三次B样条曲面
surface = Bsurface(points)
X = getReal(surface[:,0])
Y = getReal(surface[:,1])
Z = getReal(surface[:,2])
ax3.plot_surface(X,Y,Z,cmap='rainbow')

## 根据Bezier曲面的下边界可确定四个点，根据B样条曲面的上边界计算可确定四个边界点，后插入8个控制点即可获得连续的Bezier曲面


## 根据Bezier曲面的下边界可计算获得四个控制点，根据B样条曲面的上边界获取四个控制点，后插入8个控制点即可获得连续的三次B样条曲面

ax3.set_xlabel('X', fontdict={'size': 15, 'color': 'black'})
ax3.set_ylabel('Y', fontdict={'size': 15, 'color': 'black'})
ax3.set_zlabel('Z', fontdict={'size': 15, 'color': 'black'})
plt.show()
```
