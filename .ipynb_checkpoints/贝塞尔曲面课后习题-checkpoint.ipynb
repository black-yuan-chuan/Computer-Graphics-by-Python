### 已知双三次Bezier曲面P(u,v)由以下控制点定义：

$$
\begin{align}
&P_{0,0}(0,0,0),  &&P_{0,1}(20,0,20),  &&P_{0,2}(40,0,20),  &&P_{0,3}(60,0,0), \\
&P_{1,0}(0,20,20),&&P_{1,1}(20,20,40), &&P_{1,2}(40,20,40), &&P_{1,3}(60,20,20),\\
&P_{2,0}(0,40,20),&&P_{2,1}(20,40,40), &&P_{2,2}(40,40,40), &&P_{2,3}(60,40,20),\\
&P_{3,0}(0,60,0), &&P_{3,1}(20,60,20), &&P_{3,2}(40,60,20), &&P_{3,3}(60,60,0) \\
\end{align}
$$

求该Bezier曲面上的点$P(0.5,0.5)$的值。

1.输入所有的三阶Bezier曲面的控制点和锚点坐标

```python
import numpy as np
from scipy.special import comb

points = np.array([[[0,0,0], [20, 0, 20], [40, 0, 20], [60, 0,0]],\
                   [[0, 20, 20], [20, 20, 40], [40, 20, 40], [60, 20, 20]],\
                   [[0, 40, 20], [20, 40, 40], [40, 40, 40], [60, 40, 20]],\
                   [[0, 60, 0], [20, 60, 20], [40, 60, 20], [60, 60, 0]]])
```

2.通过Bezier曲面定义计算$B_{i,3}(u)$和$B_{j,3}(v)$, 下面分别用参数bu和bv指代

```python
bu = np.array([comb(3,0) * 0.5 ** 3, comb(3,1) * 0.5 ** 3, comb(3, 2) * 0.5 **3, comb(3, 3) * 0.5 ** 3])
bv = bu
```

3.曲面公式，令$U = [B_{0,3}(u), B_{1,3}(u), B_{2,3}(u), B_{3,3}(u)]$以及$V = [B_{0,3}(v), B_{1,3}(v), B_{2,3}(v), B_{3,3}(v)]$，
则$p(u,v)=UPV^T$

```python
ans = np.dot(bu, np.dot(bv, points))
print(ans)
```

## 绘制Bezier曲面



1.输入所有三阶Bezier曲面的控制点和锚点

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

points = np.array([[[0,50, 50],[100,50, 100],[100,50, 150],[0,50, 200]],\
                   [[100,100,50],[200,100,100],[200,100,150],[100,100,200]],\
                   [[100,150,50],[200,150,100],[200,150,150],[100,150,200]],\
                   [[0,200,50],[100,200,100],[100,200,150],[0,200,200]]])
```

2.以散点图的形式将控制点显示在三维画布上

```python
fig1 = plt.figure(figsize=(10,10))
ax = Axes3D(fig1)
ax.scatter(points[:,:,2], points[:,:,1], points[:,:,0], c='r', label='control points')

ax.set_xlim(0, 200)
ax.set_ylim(0, 200)
ax.set_zlim(0, 200)
ax.set_zlabel('X', fontdict={'size': 15, 'color': 'black'})
ax.set_ylabel('Y', fontdict={'size': 15, 'color': 'black'})
ax.set_xlabel('Z', fontdict={'size': 15, 'color': 'black'})
plt.show()
```

3.定义函数用户计算给定参数$u,v$后$B_{i,3}(u)$和$B_{j,3}(v)$的值, 分别用函数getBu和getBv指代；定义函数计算在$B_{i,3}(u)$和$B_{j,3}(v)$给定后根据控制点坐标信息计算曲面在欧氏空间的三维坐标,用函数puv指代。

```python
def puv(u, v, points):
    bu = np.array([comb(3,0) * (1 - u) ** 3, comb(3,1) * u * (1 - u) ** 2, comb(3, 2) * u ** 2 * (1-u), comb(3, 3) * u ** 3])
    bv = np.array([comb(3,0) * (1 - v) ** 3, comb(3,1) * v * (1 - v) ** 2, comb(3, 2) * v ** 2 * (1-v), comb(3, 3) * v ** 3])
    return np.dot(bu, np.dot(bv, points))
```

4.使用numpy中的函数构建参数$u,v$，$100 \times 100$的点阵, 将数组拉平得到10000个以u，v为坐标轴的点

```python
u = np.arange(0.0, 1.0, 0.01)
v = np.arange(0.0, 1.0, 0.01)
u, v = np.meshgrid(u, v)
u = u.flatten()
v = v.flatten()
```

5.将10000个点循环代入puv函数中计算其欧式空间坐标

```python
ans = np.array([[0,0,0]])
for i in range(10000):
    p = puv(u[i], v[i], points).reshape(1,3)
    ans = np.concatenate((ans, p), axis=0)
ans = ans[1:, :]
```

6.将上面计算的点坐标，至于3维画布中显示出来

```python
fig2 = plt.figure(figsize=(10,10))
ax3= Axes3D(fig2)
X = ans[:,0].reshape(100, 100)
Y = ans[:,1].reshape(100, 100)
Z = ans[:,2].reshape(100, 100)
ax3.plot_surface(X,Y,Z,cmap='rainbow')
ax3.scatter(points[:,:,0], points[:,:,1], points[:,:,2], c='r', label='control points')
 
ax3.set_xlim(0, 200)
ax3.set_ylim(0, 200)
ax3.set_zlim(0, 200)
ax3.set_xlabel('X', fontdict={'size': 15, 'color': 'black'})
ax3.set_ylabel('Y', fontdict={'size': 15, 'color': 'black'})
ax3.set_zlabel('Z', fontdict={'size': 15, 'color': 'black'})
plt.show()
```
