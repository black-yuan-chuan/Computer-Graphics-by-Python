## 2.NURB曲线曲面


2.1 通过NurbsDemo（NURBSpline）构造由:

- 控制点：$(0,0,50),(0,50,40),(0,75,100),(0,25,125),(0,-50,150),(0,-25,225),(0,75,250)$ 
- 节点向量：$(0,0,0,0,0.2,0.5,0.9,1,1,1,1)$
- 节点权值：$(1,0.5,1,0.8,1,0.5,1)$


```python
def countNik(points, array):
    n = len(points)
    l = len(array)
    k = l - n - 1
    return n, l, k
```

```python
points = [(0,0,50),(0,50,40),(0,75,100),(0,25,125),(0,-50,150),(0,-25,225),(0,75,250)]
array = [0,0,0,0,0.2,0.5,0.9,1,1,1,1]
print(countNik(points, array))
```

```python
countNik(points, array)
```

```python
from geomdl import NURBS

# Create a 3-dimensional B-spline Curve
curve = NURBS.Curve()
curve.degree = 3
curve.ctrlpts =  [[0,0,50],[0,50,40],[0,75,100],[0,25,125],[0,-50,150],[0,-25,225],[0,75,250]]
curve.knotvector = [0, 0, 0, 0, 0.2, 0.5, 0.9, 1, 1, 1, 1]
curve.weights = [1, 0.5, 1, 0.8, 1, 0.5, 1]
curve.delta = 0.01
curve_points = curve.evalpts
```

```python
import numpy as np
curve_points = np.array(curve_points)
```

```python
import matplotlib.pyplot as plt

fig2 = plt.figure(figsize=(10,10))
plt.scatter([points[i][2] for i in range(7)],[points[i][1] for i in range(7)], c="black")
plt.plot(curve_points[:,2], curve_points[:,1], color='red')
plt.show()
```
